### The crosstab keyword allows the user to create a view of the columns as a cross tab. The desired columns are arranged in a row and column arrangement and the comparison of the same is showcased. The columns incuded in the crosstab can be an array of columns from the dataset or individual columns based on the requirement.
### In this example we are displaying a crosstab of the home teams against the away teams. The output is the number of matches played between two teams in all World Cup matches


```python
a=wc.Home.value_counts()+wc.Away.value_counts()
label=a.index
matches=a.values
my_colors="rgbykmc"
mp.figure(figsize=(12,7))

mp.bar(label,matches,align='center',color=my_colors)

mp.title("Number of Matches Played by Every Country in the World Cup")
mp.xlabel("Country")
mp.ylabel("Number of Matches Played")
mp.xticks(label,fontsize=5)
mp.show()
```

### Bar graph of the number of matches in each stage in the World Cup

```python
wc.isnull().sum()

#removing null values
wc = wc[wc["Year"].notnull()]

att = wc.groupby("Year")["Attendance"].sum().reset_index()
att["Year"] = att["Year"].astype(int)
mp.figure(figsize=(12,7))
sns.barplot(att["Year"],att["Attendance"],linewidth=1)
mp.grid(True)
mp.title("Attendance by year")
mp.show()
```

### Bar graph of Attendance per World Cup 

```python
ht = wc["Home Team Name"].value_counts().reset_index()
ht.columns = ["team","matches"]
at = wc["Away Team Name"].value_counts().reset_index()
at.columns = ["team","matches"]
mt = pd.concat([ht,at],axis=0)
mt = mt.groupby("team")["matches"].sum().reset_index().sort_values(by="matches",ascending=False)

mp.figure(figsize=(10,13))
ax = sns.barplot("matches","team",data=mt[:25], linewidth=1)
mp.grid(True)
mp.title("Number of matches played by teams")
for i,j in enumerate("Matches played  : " + mt["matches"][:25].astype(str)):
    ax.text(.7,i,j,fontsize=12,color="white")
```

### Boxplot of total goals scored in each FIFA World Cup

```python
import networkx as nx 

def interactions(year,color):
    
    df  =  wc[wc["Year"] == year][["Home Team Name","Away Team Name"]]
    G   = nx.from_pandas_edgelist(df,"Home Team Name","Away Team Name")
    
    mp.figure(figsize=(10,9))
    
    nx.draw_kamada_kawai(G,with_labels = True,
                         node_size  = 2500,
                         node_color = color,
                         node_shape = "h",
                         edgecolor  = "k",
                         linewidths  = 5 ,
                         font_size  = 13 ,
                         alpha=.8)
    
    mp.title("Interaction between teams :" + str(year) , fontsize =13 , color = "b")
    
```

### A function which defines interactions between each team in a given world cup

```python
interactions(2014,"r")
```

```python
interactions(1930,"r")
```
