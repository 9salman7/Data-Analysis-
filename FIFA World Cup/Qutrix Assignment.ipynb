```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Salman\Downloads\chart_work.csv")
```

```python
df.head()
```

```python
x = df.groupby("ProductName")["NumberOfBugs"].sum().reset_index()
plt.figure(figsize=(15,7))
sns.barplot(x["ProductName"],x["NumberOfBugs"],linewidth=1)
plt.title("Bar Graph of Product Name v/s Number of Bugs")
plt.savefig("BarGraph")
plt.show()
```

```python
p=["Arista Router","Cisco Router","Cisco Switch","Dell Server","HP Server","Juniper Router"]
n=[139,113,13,72,154,74]
plt.figure(figsize=(15,7))

plt.title("Pie Chart showing the distribution of number of bugs for all routers")
plt.pie(n,labels=p,explode=(0,0,0,0,0.1,0))
plt.savefig("PieChart.jpg")
```

```python
plt.figure(figsize=(15,14))
ax=plt.axes()
ax.set(xlabel="Date",ylabel="Number of Bugs")
plt.title("Linear Plot showing the distribution of number of bugs v/s Date")
plt.plot(df.Date,df.NumberOfBugs)
plt.savefig("LinearPlot.jpg")
```
